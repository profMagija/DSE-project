```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
searchall = pd.read_csv("./project_evaluation_searchall.csv", sep=';')
searchall
```

```python
searchfirst = pd.read_csv("./project_evaluation_searchfirst.csv", sep=';')
searchfirst
```

```python
plt.figure(figsize=(9,6))

plt.plot(searchall["graph size"], searchall["graph size"], marker='o', color='b', label='optimal')
plt.plot(searchall["graph size"], searchall["project search 10"], marker='o', color='orange', label='search with degree 10')
plt.plot(searchall["graph size"], searchall["project search 5"], marker='o', color='g', label='search with degree 5')
plt.plot(searchall["graph size"], searchall["original search"], marker='o', color='r', label='original search')

plt.title('Number of files found vs size of the network', fontsize=14)
plt.xlabel('Size of the network', fontsize=12)
plt.ylabel('Number of files found', fontsize=12)
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize=(9,6))

plt.plot(searchfirst["graph size"], searchfirst["project search 10"], marker='o', color='orange', label='search with degree 10')
plt.plot(searchfirst["graph size"], searchfirst["project search 5"], marker='o', color='g', label='search with degree 5')
plt.plot(searchfirst["graph size"], searchfirst["original search"], marker='o', color='r', label='original search')

plt.title('Time to find the file vs size of the network', fontsize=14)
plt.xlabel('Size of the network', fontsize=12)
plt.ylabel('Time to find the file (us)', fontsize=12)
plt.legend()
plt.grid(True)
```

```python
time_size = pd.read_csv("./project_evaluation_time_size.csv", sep=';')
time_size
```

```python
time_connection = pd.read_csv("./project_evaluation_time_connection.csv", sep=';')
time_connection
```

```python
plt.figure(figsize=(9,6))

plt.plot(time_size["NumNodes"], time_size["Average - Time Median"], marker='o', color='orange', label='Project')
plt.plot(time_size["NumNodes"], time_size["Average - Time2 Median"], marker='o', color='g', label='Original')

plt.title('Median download time vs size of the network', fontsize=14)
plt.xlabel('Number of nodes', fontsize=12)
plt.ylabel('Median download time [ms]', fontsize=12)
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize=(9,6))

plt.plot(time_connection["NumConns"], time_connection["Average - Time Median"], marker='o', color='orange', label='Project')
plt.plot(time_connection["NumConns"], time_connection["Average - Time2 Median"], marker='o', color='g', label='Original')

plt.title('Median download time vs number of connections per node', fontsize=14)
plt.xlabel('Connections per node', fontsize=12)
plt.ylabel('Median download time [ms]', fontsize=12)
plt.ylim((0,300))
plt.legend()
plt.grid(True)
```
